## 1. What is NumPy

NumPy is a package for multi-dimensional arrays. 
It contains basic functions to work with arrays, but also contains complex modules designed to do very specific things (see below).




### Whye are we using NumPy instead of lists?

An array is a collection of elements of the **same type**. This makes arrays less flexible than lists, however, arrays are designed to be multi-dimensional, making them a powerful way to store geoscientific data (e.g. can represent temporal and spatial axes). 

Furthermore computation with NumPy arrays is much faster. This has several reasons:

- NumPy arrays are collections of a fixed data type. The data type information has not to be checked before each operation
- NumPy arrays are using contiguous memory location
- NumPy breaks down a task into fragments and then processes it all in parallel. 
- NumPy integrates C, C++, and Fortran code in Python. These languages are very fast.


For additional resources on numpy, there are excellent tutorials on the official webpage (https://numpy.org/doc/stable/user/absolute_beginners.html), on youtube (https://www.youtube.com/results?search_query=introduction+numpy) and many other places on the internet. If you have a specific question on a function check out the official help, but often, stackoverflow is a valuable resource as well.

### Interpolation

NumPy has a function for simple linear interpolation. 
 'numpy.interp(x, xp, fp)'

```python
#create some x and y data
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)

xvals = np.linspace(0, 2*np.pi, 50) #points where we want to interpolate to
#interpolation
yinterp = np.interp(xvals, x, y)
#plot 
plt.plot(x, y, 'o')
plt.plot(xvals, yinterp, '-x')
plt.show()
```

### Using Scipy for linear regression

This is an example how we can use Scipy to calculate a linear regression of two data sets. It automatically calculates slope, intercept, rvalue, pvalue, stderr, intercept_stderr. 

```python
from scipy import stats

# generate some data 
x = np.random.rand(10)
y = 1.6*x + np.random.rand(10)

#Perform linear regression
res = stats.linregress(x,y)
```

```python
#Access values from the regression model
print(res.rvalue)
```

```python
#plot data
plt.plot(x, y, 'o', label='original data')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.legend()
plt.show()
```

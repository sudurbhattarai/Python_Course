## 1. What is NumPy

NumPy is a package for multi-dimensional arrays. 
It contains basic functions to work with arrays, but also contains complex modules designed to do very specific things (see below).




### Whye are we using NumPy instead of lists?

An array is a collection of elements of the **same type**. This makes arrays less flexible than lists, however, arrays are designed to be multi-dimensional, making them a powerful way to store geoscientific data (e.g. can represent temporal and spatial axes). 

Furthermore computation with NumPy arrays is much faster. This has several reasons:

- NumPy arrays are collections of a fixed data type. The data type information has not to be checked before each operation
- NumPy arrays are using contiguous memory location
- NumPy breaks down a task into fragments and then processes it all in parallel. 
- NumPy integrates C, C++, and Fortran code in Python. These languages are very fast.


For additional resources on numpy, there are excellent tutorials on the official webpage (https://numpy.org/doc/stable/user/absolute_beginners.html), on youtube (https://www.youtube.com/results?search_query=introduction+numpy) and many other places on the internet. If you have a specific question on a function check out the official help, but often, stackoverflow is a valuable resource as well.

### Interpolation

NumPy has a function for simple linear interpolation. 
 'numpy.interp(x, xp, fp)'

```python
#create some x and y data
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)

xvals = np.linspace(0, 2*np.pi, 50) #points where we want to interpolate to
#interpolation
yinterp = np.interp(xvals, x, y)
#plot 
plt.plot(x, y, 'o')
plt.plot(xvals, yinterp, '-x')
plt.show()
```

### Using Scipy for linear regression

This is an example how we can use Scipy to calculate a linear regression of two data sets. It automatically calculates slope, intercept, rvalue, pvalue, stderr, intercept_stderr. 

```python
from scipy import stats

# generate some data 
x = np.random.rand(10)
y = 1.6*x + np.random.rand(10)

#Perform linear regression
res = stats.linregress(x,y)
```

```python
#Access values from the regression model
print(res.rvalue)
```

```python
#plot data
plt.plot(x, y, 'o', label='original data')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.legend()
plt.show()
```

## check time and convert to a commonly used date format (2 Points)

Date-time is in Julian format, the commonly used format is Gregorian, so the JULIAN date should be converted to GREGORIAN date format.

```python
def julian_to_gregorian(julian_dates):
    # Add the offset between Julian and Gregorian calendar
    gregorian_offset = 2400000.5
    # Calculate the number of days and fractions of a day
    days = julian_dates - gregorian_offset
    # Calculate the datetime components
    hours = (days * 24) % 24
    minutes = (hours * 60) % 60
    seconds = (minutes * 60) % 60
    # Convert days to datetime objects
    dates = np.vectorize(datetime.datetime.fromordinal)(np.floor(days).astype(int) + datetime.datetime(1858, 11, 17).toordinal())
    # Extract Gregorian dates and times
    time = [date.date() for date in dates]
    gregorian_times = [datetime.time(h, m, s) for h, m, s in zip(hours.astype(int), minutes.astype(int), seconds.astype(int))]
    return time, gregorian_times

# Load the NetCDF file
dataset = Dataset(r'D:/MSRSGI/Winter_Semester23_24/Python/Course_Data/GT_MDB_IS_INSIT_KIT_F-20211231_2331XX-KKIT-0.00X0.00-V.1.0.nc')

# Read the Julian date variable from the NetCDF file
julian_date_variable = dataset.variables['jul_date'][:600000]  # Select the first 1000 dates

# Convert Julian dates to Gregorian dates
time, gregorian_times = julian_to_gregorian(julian_date_variable)

# Prepare the data for CSV writing
data = list(zip(time, gregorian_times))

# Specify the directory to save the CSV file
output_directory = 'D:/MSRSGI/Winter_Semester23_24/Python/Course_Data/'

# Ensure the output directory exists, create it if necessary
os.makedirs(output_directory, exist_ok=True)

# Specify the CSV file path
csv_file = os.path.join(output_directory, 'KIT_F_time_Gregorian.csv')

# Write data to the CSV file
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['time', 'Gregorian_Time'])  # Write header
    writer.writerows(data)  # Write data rows

print("Data saved to CSV file:",csv_file)
```

## 2. Scatterplot of temperature vs. relative humidity with a regression line, and visualize the regression equation and the r-squared value in your plot (2 Points)

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#defining x and y values
x = df_final.airtemp200m
y = df_final.relativehumidity200m
```

```python
#perform linear regression
res = stats.linregress(x, y)
```

```python
#Acess the values  from the regression model
slope  = res.slope
intercept = res.intercept
r_value = res.rvalue #gives the correlation coefficient
r_squared = r_value**2 #measaure of coefficent of deteremination, square of correlation coefficent; measure of how well the regression line fits the observed data
```

```python
#Plot data and regression line
plt.scatter(x, y, label = 'Original Data')

plt.plot(x, intercept + slope * x, 'r', label=f'Fitted Line\ny = {slope:.2f}x + {intercept:.2f}\nR-squared = {r_squared:.2f}')

print(f'y = {slope:.2f}x + {intercept:.2f}\nR-squared = {r_squared:.2f}')

plt.legend()
plt.xlabel('200m Air Temperature')
plt.ylabel('200m Relative Humidity')
plt.title('Air Temperature vs. Relative Humidity')
plt.show()
```

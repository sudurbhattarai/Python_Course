```python
# Read in data as pandas dataframe
df = pd.read_csv('GIS_GFZOP_0600_0003.csv', index_col='time [yyyy-mm-dd]', parse_dates=True)
df
```

```python
# Plotting mean values over time
plt.figure(figsize=(24, 6))

plt.plot(df.index, df['global [Gt]'], 'ro-', label="global", linewidth = '2')
plt.plot(df.index, df['uncertainty_global [Gt]'], 'bo-', label="global uncertainty, linewidth = '2')


plt.title('Daily Mean Ov')
plt.xlabel('TIMESTAMP')
plt.ylabel('Mean Values')
plt.legend()

# Save the figure
plt.savefig('Daily_Mean_Over_Time.png')

plt.show()
```
